# **Stressed PD Model Illustration**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The model: pd_growth=a+rho*pd_growth_lag+b1*x1+b2*x2
a=0.01
rho=0.5
pd=0.05 # pd at time t
pdl1=0.04 # pd at time t-1
pdl2=0.03 # pd at time t-2
pd_growth=pd/pdl1-1 # pd growth rate at time t
pd_growth_lag=pdl1/pdl2-1 # pd growth rate at time t-1
x1=0.02 # gdp growth rate at time t, constant forecast
x2=0.01 # term spread at time t, constant forecast
```

```python
# Define the coefficients
a1= 0
b1 = 0
c1 = 1

# Define the pd values in the interval [0, 1]
pd1 = np.linspace(0, 1, 100)

# Calculate b2 for each pd value
b1pd = a1 + b1 * pd1 + c1 * pd1**2

# Plot the function
plt.plot(pd1, b1pd, label='b1pd = a1 + b1 * pd1 + c1 * pd1^2')
plt.xlabel('pd1')
plt.ylabel('b1pd')
plt.title('Quadratic Function b1pd = a1 + b1 * pd1 + c1 * pd1^2')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the coefficients
a2 = 0
b2 = 1
c2 = -1

# Define the pd values in the interval [0, 1]
pd2 = np.linspace(0, 1, 100)

# Calculate b2 for each pd value
b2pd = a2 + b2 * pd2 + c2 * pd2**2

# Plot the function
plt.plot(pd, b2pd, label='b2pd = a2 + b2 * pd2 + c2 * pd2**2')
plt.xlabel('pd')
plt.ylabel('b2pd')
plt.title('Quadratic Function b2pd = a2 + b2 * pd2 + c2 * pd2**2')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
